**Multivariate Analysis (Higher-Dimensional Data Relationships)**<br>
Objective: Analyze relationships in higher-dimensional data using pair plots and PCA.

Title: Pair Plots<br>

Task 1: Create a pair plot for the Iris dataset.<br>
Task 2: Focus on a subset of features (e.g., only petal dimensions).<br>
Task 3: Exclude one class to observe differences.

```python
# Write your code from here
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Rename columns for easier handling
df.rename(columns={
    'sepal length (cm)': 'sepal_length',
    'sepal width (cm)': 'sepal_width',
    'petal length (cm)': 'petal_length',
    'petal width (cm)': 'petal_width'
}, inplace=True)

# Task 1: Pair plot for the full Iris dataset
sns.pairplot(df, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

# Task 2: Pair plot focusing on petal dimensions only
sns.pairplot(df, vars=['petal_length', 'petal_width'], hue='species')
plt.suptitle('Pair Plot of Petal Dimensions', y=1.02)
plt.show()

# Task 3: Pair plot excluding one class (e.g., exclude 'setosa')
df_no_setosa = df[df['species'] != 'setosa']
sns.pairplot(df_no_setosa, hue='species')
plt.suptitle('Pair Plot Excluding Setosa', y=1.02)
plt.show()
```

Title: Principal Component Analysis (PCA)<br>

Task 1: Perform PCA and plot the first two principal components.<br>
Task 2: Visualize explained variance by each component.<br>
Task 3: Retain more components and visualize in 3D.

```python
# Write your code from here
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting
import seaborn as sns

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Rename columns for convenience
df.rename(columns={
    'sepal length (cm)': 'sepal_length',
    'sepal width (cm)': 'sepal_width',
    'petal length (cm)': 'petal_length',
    'petal width (cm)': 'petal_width'
}, inplace=True)

# Features for PCA
X = df.drop('species', axis=1)

# Task 1: Perform PCA and plot the first two components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['species'], palette='Set1', s=80)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: First Two Principal Components')
plt.legend()
plt.show()

# Task 2: Visualize explained variance ratio
plt.figure(figsize=(8, 4))
plt.bar(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, alpha=0.7)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components (2 Components)')
plt.show()

# Task 3: Retain more components and visualize in 3D
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2],
    c=iris.target, cmap='Set1', s=80, edgecolor='k'
)
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
ax.set_title('PCA: First Three Principal Components')
legend1 = ax.legend(*scatter.legend_elements(), title="Species")
ax.add_artist(legend1)
plt.show()
```

**Statistical Analysis in EDA**<br>

Objective: Calculate basic statistical metrics and explore the relationship between features using correlation and covariance.<br>

Title: Descriptive Statistics<br>

Task 1: Calculate mean, median, and standard deviation of petal length.<br>
Task 2: Calculate skewness and kurtosis of sepal width.<br>
Task 3: Calculate mean, median, and standard deviation of sepal length.

```python
# Write your code from here
import pandas as pd
from sklearn.datasets import load_iris
from scipy.stats import skew, kurtosis

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Rename columns for convenience
df.rename(columns={
    'sepal length (cm)': 'sepal_length',
    'sepal width (cm)': 'sepal_width',
    'petal length (cm)': 'petal_length',
    'petal width (cm)': 'petal_width'
}, inplace=True)

# Task 1: Mean, median, std of petal length
petal_length_mean = df['petal_length'].mean()
petal_length_median = df['petal_length'].median()
petal_length_std = df['petal_length'].std()

print(f"Petal Length - Mean: {petal_length_mean:.3f}, Median: {petal_length_median:.3f}, Std Dev: {petal_length_std:.3f}")

# Task 2: Skewness and kurtosis of sepal width
sepal_width_skewness = skew(df['sepal_width'])
sepal_width_kurtosis = kurtosis(df['sepal_width'])

print(f"Sepal Width - Skewness: {sepal_width_skewness:.3f}, Kurtosis: {sepal_width_kurtosis:.3f}")

# Task 3: Mean, median, std of sepal length
sepal_length_mean = df['sepal_length'].mean()
sepal_length_median = df['sepal_length'].median()
sepal_length_std = df['sepal_length'].std()

print(f"Sepal Length - Mean: {sepal_length_mean:.3f}, Median: {sepal_length_median:.3f}, Std Dev: {sepal_length_std:.3f}")
```

Title: Correlation & Covariance<br>

Task 1: Compute correlation between sepal length and petal length.<br>
Task 2: Compute covariance between petal width and sepal width.<br>
Task 3: Determine the most correlated pair of features.<br>

```python
# Write your code from here
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Rename columns for convenience
df.rename(columns={
    'sepal length (cm)': 'sepal_length',
    'sepal width (cm)': 'sepal_width',
    'petal length (cm)': 'petal_length',
    'petal width (cm)': 'petal_width'
}, inplace=True)

# Task 1: Correlation between sepal length and petal length
corr_sepal_petal_length = df['sepal_length'].corr(df['petal_length'])
print(f"Correlation (sepal length, petal length): {corr_sepal_petal_length:.3f}")

# Task 2: Covariance between petal width and sepal width
cov_petal_sepal_width = df['petal_width'].cov(df['sepal_width'])
print(f"Covariance (petal width, sepal width): {cov_petal_sepal_width:.3f}")

# Task 3: Find the most correlated pair of features
corr_matrix = df.corr().abs()  # absolute values to find strongest correlation regardless of sign

# Mask diagonal (correlation of feature with itself)
for i in range(len(corr_matrix)):
    corr_matrix.iat[i, i] = 0

# Find max correlation value and corresponding pair
max_corr = corr_matrix.unstack().idxmax()
max_value = corr_matrix.unstack().max()

print(f"Most correlated pair: {max_corr} with correlation of {max_value:.3f}")
```
