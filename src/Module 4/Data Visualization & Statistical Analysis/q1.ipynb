Title: Histogram of a Single Feature <br>

Question 1: Create a histogram for the age feature from a dataset. Interpret what the shape of the histogram tells us about the distribution of the age feature.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic age data (e.g., 500 people aged 18-70)
np.random.seed(42)
age_data = np.random.normal(loc=35, scale=10, size=500)
age_data = np.clip(age_data, 18, 70)  # constrain ages between 18 and 70

df = pd.DataFrame({'age': age_data})

# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=False, color='skyblue')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Title: Boxplot for a Single Feature <br>

Question 2: Generate a boxplot for the salary feature and identify any outliers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create synthetic salary data with outliers
np.random.seed(0)
salary_data = np.concatenate([
    np.random.normal(50000, 8000, 300),   # Regular salaries
    np.random.normal(120000, 5000, 5)     # High salary outliers
])

df = pd.DataFrame({'salary': salary_data})

# Plot the boxplot
plt.figure(figsize=(6, 8))
sns.boxplot(y=df['salary'], color='lightcoral')
plt.title('Boxplot of Salary')
plt.ylabel('Salary ($)')
plt.show()

# Identify outliers programmatically
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['salary'] < (Q1 - 1.5 * IQR)) | (df['salary'] > (Q3 + 1.5 * IQR))]

print("Outlier salary values:")
print(outliers['salary'].values)
```

Title: Correlation Heatmap<br>

Question 5 : Generate a correlation heatmap for a dataset with multiple features (e.g., height ,weight , age ) and explain the correlations observed.

```python
# Write your code from here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data
np.random.seed(42)
height = np.random.normal(170, 10, 300)  # cm
weight = 0.5 * height + np.random.normal(0, 8, 300)  # kg, correlated with height
age = np.random.normal(40, 12, 300)  # years

df = pd.DataFrame({'height': height, 'weight': weight, 'age': age})

# Compute correlation matrix
corr = df.corr()

# Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

Title: Descriptive Statistical Analysis<br>

Question 6: Calculate the mean, median, standard deviation, skewness, and kurtosis for the
temperature feature and discuss the results.

```python
# Write your code from here
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

# Generate synthetic temperature data (in °C)
np.random.seed(123)
temperature_data = np.random.normal(loc=20, scale=5, size=500)  # average 20°C, std dev 5°C
temperature_data = np.clip(temperature_data, -5, 40)  # realistic temperature bounds

df = pd.DataFrame({'temperature': temperature_data})

# Calculate descriptive statistics
mean_temp = df['temperature'].mean()
median_temp = df['temperature'].median()
std_temp = df['temperature'].std()
skew_temp = skew(df['temperature'])
kurt_temp = kurtosis(df['temperature'])

print(f"Mean temperature: {mean_temp:.2f} °C")
print(f"Median temperature: {median_temp:.2f} °C")
print(f"Standard deviation: {std_temp:.2f}")
print(f"Skewness: {skew_temp:.2f}")
print(f"Kurtosis: {kurt_temp:.2f}")
```

Title: Covariance and Correlation between Two Features<br>

Question 7: Compute the covariance and correlation between price and demand in a dataset.
Explain what these metrics indicate.

```python
# Write your code from here
import numpy as np
import pandas as pd

# Generate synthetic data for price and demand
np.random.seed(42)
price = np.random.normal(50, 10, 300)  # average price $50
# Demand negatively correlated with price: as price rises, demand falls
demand = 1000 - 10 * price + np.random.normal(0, 30, 300)

df = pd.DataFrame({'price': price, 'demand': demand})

# Compute covariance
cov_matrix = df.cov()
cov_price_demand = cov_matrix.loc['price', 'demand']

# Compute correlation
corr_matrix = df.corr()
corr_price_demand = corr_matrix.loc['price', 'demand']

print(f"Covariance between price and demand: {cov_price_demand:.2f}")
print(f"Correlation between price and demand: {corr_price_demand:.2f}")
```

Title: Principal Component Analysis (PCA)<br>

Question 9 : Apply PCA on a dataset with multiple features (e.g., x1 , x2 , x3 , x4 ) and reduce it to two principal components. Visualize the data in the new feature space.

```python
# Write your code from here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Generate synthetic data with 4 features
np.random.seed(42)
x1 = np.random.normal(0, 1, 300)
x2 = 0.8 * x1 + np.random.normal(0, 0.5, 300)  # correlated with x1
x3 = np.random.normal(0, 1, 300)
x4 = 0.3 * x3 + np.random.normal(0, 0.7, 300)  # correlated with x3

df = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4})

# Standardize features before PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

# Create DataFrame of principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Plot the principal components
plt.figure(figsize=(8,6))
plt.scatter(pc_df['PC1'], pc_df['PC2'], alpha=0.6, color='purple')
plt.title('PCA: Projection onto First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# Optional: Explained variance ratio
print("Explained variance ratio of PC1 and PC2:", pca.explained_variance_ratio_)
```

Title: Advanced Pair Plot with Hue Parameter<br>

Question 10 : Create a pair plot for height , weight , and age with an added categorical variable gender as the hue to observe different group trends.

```python
# Write your code from here
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
size = 300

# Features
height = np.random.normal(170, 10, size)
weight = 0.5 * height + np.random.normal(0, 8, size)
age = np.random.normal(40, 12, size)

# Categorical variable 'gender'
genders = np.random.choice(['Male', 'Female'], size=size)

df = pd.DataFrame({
    'height': height,
    'weight': weight,
    'age': age,
    'gender': genders
})

# Create pair plot with hue
sns.pairplot(df, hue='gender', palette='Set2', diag_kind='kde')
plt.suptitle('Pair Plot of Height, Weight, and Age by Gender', y=1.02)
plt.show()

```
